```python
df = df.drop_duplicates()

date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

df['order_status'] = df['order_status'].str.strip().str.lower()

df = df.dropna(subset=['order_id', 'customer_id'])

df['delivery_delay_days'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days
```

```python
df.head()

df.to_csv('../data/cleaned_orders_dataset.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('../data/olist_orders_dataset.csv')

# remove exact duplicates
df.drop_duplicates(inplace=True)

# convert datetime columns safely
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# standardize text
df['order_status'] = df['order_status'].str.strip().str.lower()

# DO NOT drop rows here (important)

# optional derived column (keep for analysis, ignore in SQL load)
df['delivery_delay_days'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days

df.to_csv('../data/cleaned_orders_dataset.csv', index=False)
```

```python
df = pd.read_csv('../data/olist_customers_dataset.csv')

df.drop_duplicates(inplace=True)

# standardize location
df['customer_city'] = df['customer_city'].str.strip().str.lower()
df['customer_state'] = df['customer_state'].str.strip().str.upper()

# DO NOT drop rows

df.to_csv('../data/cleaned_customers_dataset.csv', index=False)
```

```python
df = pd.read_csv('../data/olist_payments_dataset.csv')

df.drop_duplicates(inplace=True)

# clean categorical
df['payment_type'] = df['payment_type'].str.strip().str.lower()

# ensure numeric
df['payment_value'] = pd.to_numeric(df['payment_value'], errors='coerce')
df['payment_installments'] = pd.to_numeric(df['payment_installments'], errors='coerce')

df.to_csv('../data/cleaned_payments_dataset.csv', index=False)
```

```python
df = pd.read_csv('../data/olist_order_items_dataset.csv')

df.drop_duplicates(inplace=True)

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['freight_value'] = pd.to_numeric(df['freight_value'], errors='coerce')

# optional: convert shipping date
df['shipping_limit_date'] = pd.to_datetime(df['shipping_limit_date'], errors='coerce')

df.to_csv('../data/cleaned_order_items_dataset.csv', index=False)
```

```python
df = pd.read_csv('../data/olist_products_dataset.csv')

df.drop_duplicates(inplace=True)

df['product_category_name'] = df['product_category_name'].str.strip().str.lower()

# keep only needed columns for SQL
df = df[['product_id', 'product_category_name', 'product_weight_g']]

df.to_csv('../data/cleaned_products_dataset.csv', index=False)
```
